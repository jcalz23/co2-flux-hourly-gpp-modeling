```python
root_dir =  os.path.abspath(os.getcwd()+"..\\..\\..\\..\\")
tmp_dir = root_dir + "\\.tmp"
data_dir = root_dir + "\\data\\"
site_metadata_filename = data_dir + "hourly-site-summary.csv"
site_summary_filename = data_dir + "datasets\\site_summary_v0_3_2.csv"
final_site_metadata_filename = data_dir + "site-metadata.csv"
```

# Load site metadata generated from the data files

```python
metadata_df = pd.read_csv(site_metadata_filename, index_col='site_id').drop('index', axis=1)
print(f"size:{metadata_df.shape}")
metadata_df.tail()
```

# Load site summary from file

```python
col_names = ["site_id", "dataset", "start_year", "end_year", "file", "is_dup", "IGBP",
             "elevation", "lat", "long", "site_name", "koppen_sub", "koppen_main",
             "koppen_name", "koppen_main_name","c3c4", "c4_percent"]
summary_df = pd.read_csv(site_summary_filename, names=col_names, header=0)
summary_df['file'] = [x.split('/')[-1] for x in summary_df['file']]
print(f"size:{summary_df.shape}")
summary_df.head()
```

```python
final_df = summary_df.set_index('site_id').join(metadata_df)
final_df['start_time'] = pd.to_datetime(final_df['start_time'])
final_df['end_time'] = pd.to_datetime(final_df['end_time'])
print(f"size:{final_df.shape}")
final_df.head()
```

# Site Metadata EDA

```python
#Load site metadata
final_df = pd.read_csv(final_site_metadata_filename)
final_df['start_time'] = pd.to_datetime(final_df['start_time'])
final_df['end_time'] = pd.to_datetime(final_df['end_time'])
final_df.head()
```

```python
final_df.sort_values("end_time", ascending=False)[['site_id','size','record_count','recorded_date_count','start_time','end_time']]
```

```python
# Plot World Map
import plotly.express as px
import plotly.graph_objects as go

# Prep Data
final_df.recorded_date_count = final_df.recorded_date_count.fillna(0)
plot_df = final_df.sort_values('recorded_date_count', ascending=False)
plot_df['start'] = plot_df['start_time'].dt.date
plot_df['end'] = plot_df['end_time'].dt.date

# Plot
fig = px.scatter_geo(plot_df, 
                     lat = "lat", lon = "long",
                     size="recorded_date_count",
                     color="recorded_date_count",
                     hover_name="site_id", opacity=0.5,
                     #hover_data = {"lat":False, "long":False,
                     #              "site_name":True, "recorded_date_count": True},
                     # projection="orthographic", # 3D Earth
                     #projection = "natural earth",
                     labels={ "recorded_date_count": "Recorded<br>Days",
                              "site_name": "Site Name"})
fig.update_layout(
    title={
        'text': "Number of Recorded Days per Flux Tower",
        'y':0.9,
        'x':0},
    geo = go.layout.Geo(
        resolution = 50,
        scope = "world",
        showframe = False,
        showcoastlines = True,
        landcolor = "rgb(229, 229, 229)",
        countrycolor = "white" ,
        coastlinecolor = "white"
    )
)
fig.update_geos(lataxis_range=[-60,90])
fig.update_traces(customdata=plot_df[["site_name", "site_id", "recorded_date_count",'start','end']])
fig.update_traces(hovertemplate = "<b>%{customdata[0]} (%{customdata[1]})</b>" +
                                  "<br>Recorded Days: %{customdata[2]}" + 
                                  "<br>%{customdata[3]} ~ %{customdata[4]}")
fig.update_layout(margin={"r":0,"t":0,"l":0.5,"b":0}, height = 400, width = 700)
fig.update_layout(coloraxis_colorbar=dict(
    lenmode="pixels", len=250,
    thicknessmode="pixels", thickness=25
))

fig.show()
```
